```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = np.cumsum(np.random.randn(200))
```

```python
x = np.linspace(0, 300, 200)
x.shape
```

```python
azul = '#A1C7E0'
vermelho = '#F26363'
```

```python
fig, ax = plt.subplots(figsize=(12,5))
plt.plot(x, y)
ax.scatter(132, -22.5, s=700, color=azul)
ax.scatter(x=290, y=1.3, s=700, color=vermelho)
```
